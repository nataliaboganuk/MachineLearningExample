```python
# example model
# 4. Train and save a model
# ---------------------------------------------------------

from sklearn import svm, datasets
from joblib import dump

iris = datasets.load_iris()
X, y = iris.data, iris.target

clf = svm.SVC(gamma="scale")
clf.fit(X, y)

dump(clf, "my-model/model.joblib")

print("Model saved to my-model/model.joblib")

```

```python
#5. Start the local SKLearnServer
# ---------------------------------------------------------
python -m sklearnserver --model_dir my-model --http_port 8080

# launches a local inference server identical to the one KServe runs in the cluster.
INFO Loaded model from my-model/model.joblib
INFO Listening on http://0.0.0.0:8080

```

```python
#Send a test inference request (Jupyter)
curl -X POST \
  -H "Content-Type: application/json" \
  http://localhost:8080/v2/models/model/infer \
  -d '{
    "inputs": [
      {
        "name": "input-0",
        "shape": [1, 4],
        "datatype": "FP32",
        "data": [[5.1, 3.5, 1.4, 0.2]]
      }
    ]
  }'
```
